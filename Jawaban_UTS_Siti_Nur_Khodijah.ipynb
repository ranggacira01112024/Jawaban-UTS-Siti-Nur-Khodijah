```python
import plotly.graph_objects as go
import plotly.express as px

# Data for the charts
data = [
    {'name': 'Jan', 'value': 400},
    {'name': 'Feb', 'value': 300},
    {'name': 'Mar', 'value': 600},
    {'name': 'Apr', 'value': 200},
    {'name': 'Mei', 'value': 500}
]

pie_data = [
    {'name': 'A', 'value': 400},
    {'name': 'B', 'value': 300},
    {'name': 'C', 'value': 300},
    {'name': 'D', 'value': 200}
]

# Create Bar Chart (Vertical)
bar_chart = go.Figure()
bar_chart.add_trace(go.Bar(
    x=[item['name'] for item in data],
    y=[item['value'] for item in data],
    name='Value',
    marker_color='#8884d8'
))
bar_chart.update_layout(
    title='1. Column Chart',
    xaxis_title='Month',
    yaxis_title='Value'
)

# Create Line Chart
line_chart = go.Figure()
line_chart.add_trace(go.Scatter(
    x=[item['name'] for item in data],
    y=[item['value'] for item in data],
    mode='lines+markers',
    name='Value',
    line=dict(color='#8884d8')
))
line_chart.update_layout(
    title='2. Line Chart',
    xaxis_title='Month',
    yaxis_title='Value'
)

# Create Pie Chart
pie_chart = px.pie(
    names=[item['name'] for item in pie_data],
    values=[item['value'] for item in pie_data],
    title="3. Pie Chart",
    color_discrete_sequence=['#0088FE', '#00C49F', '#FFBB28', '#FF8042']
)

# Create Bar Chart (Horizontal)
horizontal_bar_chart = go.Figure()
horizontal_bar_chart.add_trace(go.Bar(
    x=[item['value'] for item in data],
    y=[item['name'] for item in data],
    name='Value',
    orientation='h',
    marker_color='#8884d8'
))
horizontal_bar_chart.update_layout(
    title='4. Horizontal Bar Chart',
    xaxis_title='Value',
    yaxis_title='Month'
)

# Show the charts
bar_chart.show()
line_chart.show()
pie_chart.show()
horizontal_bar_chart.show()
```
